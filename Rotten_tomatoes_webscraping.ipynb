```python
import os
import time
import requests
from bs4 import BeautifulSoup
import urllib
from selenium import webdriver 
```

```python
browser = webdriver.Firefox()
url = "https://www.rottentomatoes.com/m/ant_man_and_the_wasp"
browser.get(url)
html = browser.page_source
soup = BeautifulSoup(html,"html.parser")
```

```python
rating = soup.find('span',{'class':'meter-value superPageFontColor'} )
div = soup.find('div',{'class':'quote_bubble top_critic pull-left cl'})
review = div.find('p')
reviewer_name = div.find('a',{'class':'unstyled articleLink'})
journal = div.find('a',{'class':"small subtle articleLink"})
print rating.text
print review.text
print " "
print reviewer_name.text
print " "
print journal.text
```
